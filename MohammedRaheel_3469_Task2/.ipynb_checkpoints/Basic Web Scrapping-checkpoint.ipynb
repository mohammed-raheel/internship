```python
import requests
import bs4
import urllib.request as urlib 
```

Read wikipedia page
STEPS:
1) Use requests to make a call to URL
2) Analyze the page
3) Check which data required. In the below example we want to get the headings
4) Understand which tags need to be extracted
5) Use Beautiful soap and parse the response

```python
url = 'https://en.wikipedia.org/wiki/Machine_learning'
wikiResponse = requests.get(url)
print('URL response status: ',wikiResponse.status_code)
```

```python
# To check the content of the response, call content attribute
wikiResponse.content
```

```python
# As we can see from above, the complete HTML page is returned. Now we need to extract content that we need
# Lets say we want to extract text within the content section
# On close examination of page, we find the text we are looking for is in a list, that belongs to class 'toc'

# Create a Beautiful soap object
soup = bs4.BeautifulSoup(wikiResponse.text,'html.parser')
```

```python
# Lets extract the title of the Page
wikiTitle = soup.select('title')
print(type(wikiTitle))
print('Wiki page title: ',wikiTitle[0].getText())
```

```python
# Lets extract the text in the contents sections
wikiContents = soup.select('.toctext')
for i in wikiContents:
    print(i.text)
```

```python
# Now lets extract heading from page
wikiHeadings = soup.select('.mw-headline')
for i in wikiHeadings:
    print(i.text)
```
